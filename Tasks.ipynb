**Build a Simple “ATM Simulation” using ONLY If-Else**

```python
balance = 5000.0

print("Welcome to the ATM!")
print("1. Check Balance")
print("2. Deposit")
print("3. Withdraw")
print("4. Exit")

choice = int(input("Enter your choice (1/2/3/4): "))

if choice == 1:
    print(f"Your balance is: ₹{balance:.2f}")
elif choice in [2, 3]:
    amount = float(input("Enter amount: "))
    if amount > 0:
        if choice == 2:
            balance += amount
            print(f"Deposit successful. New balance: ₹{balance:.2f}")
        elif balance >= amount:
            balance -= amount
            print(f"Withdrawal successful. New balance: ₹{balance:.2f}")
        else:
            print("Insufficient balance.")
    else:
        print("Invalid amount.")
elif choice == 4:
    print("Thank you for using the ATM. Goodbye!")
else:
    print("Invalid choice.")
```

**Elevator Stop Decision**

```python
def elevator_stop_decision(current_floor, requested_floor, doors_open):
    if doors_open and current_floor == requested_floor:
        return "Open doors"
    elif current_floor < requested_floor:
        return "Move up"
    elif current_floor > requested_floor:
        return "Move down"
    else:
        return "Stay"

# Example usage:
print(elevator_stop_decision(5, 5, True))
print(elevator_stop_decision(3, 7, False))
print(elevator_stop_decision(8, 2, False))
print(elevator_stop_decision(4, 4, False))
```

**Smart Thermostat Heating/Cooling**

```python
def smart_thermostat(current_temp, desired_temp, is_home):
    if not is_home:
        return "Eco"
    elif current_temp < desired_temp - 0.5:
        return "Heat"
    elif current_temp > desired_temp + 0.5:
        return "Cool"
    else:
        return "Hold"

# Example usage:
print(smart_thermostat(20, 22, True))  # Output: Heat
print(smart_thermostat(25, 22, True))  # Output: Cool
print(smart_thermostat(22, 22, True))  # Output: Hold
print(smart_thermostat(20, 22, False)) # Output: Eco
```

**Parking fee calculator**

```python
import math

def parking_fee(minutes_parked, is_weekend):
    if is_weekend and minutes_parked > 120:
        return 50
    elif minutes_parked <= 30:
        return 0
    else:
        hours = math.ceil((minutes_parked - 30) / 60)
        return hours * 20

# Example usage:
print(parking_fee(20, False))  # Output: 0
print(parking_fee(60, False))  # Output: 20
print(parking_fee(150, True))  # Output: 50
```

**Printer toner warning**

```python
def printer_toner_warning(pages_printed_since_refill, printer_model_capacity):
    usage_percent = (pages_printed_since_refill / printer_model_capacity) * 100
    if usage_percent >= 95:
        return "Critical"
    elif usage_percent >= 75:
        return "Low"
    else:
        return "OK"

# Example usage:
print(printer_toner_warning(750,1000))  # Output: Low
print(printer_toner_warning(950, 1000))  # Output: Critical
print(printer_toner_warning(500, 1000))  # Output: OK
```

**Online exam proctoring lock**

```python
def online_exam_proctoring(tabs_open, webcam_on, time_into_exam_minutes, total_exam_minutes):
    if not webcam_on or tabs_open > 3:
        if (not webcam_on and tabs_open > 3) or time_into_exam_minutes > total_exam_minutes - 5:
            return "Lock exam"
        else:
            return "Warn"
    else:
        return "Continue"

# Example usage:
print(online_exam_proctoring(2, True, 30, 60))  # Output: Continue
print(online_exam_proctoring(4, True, 30, 60))  # Output: Warn
print(online_exam_proctoring(2, False, 55, 60)) # Output: Lock exam
```

**Room light color mode**

```python
def room_light_color(activity, hour):
    priorities = {"sleep": 4, "party": 3, "reading": 2, "work": 1}
    modes = {
        "reading": "Warm" if 18 <= hour <= 22 else None,
        "work": "Cool" if 9 <= hour <= 17 else None,
        "party": "Colorful" if 18 <= hour <= 23 else None,
        "sleep": "Dim" if (hour >= 22 or hour <= 6) else None
    }
    chosen_mode = None
    chosen_priority = 0
    for act, mode in modes.items():
        if mode and priorities[act] > chosen_priority:
            chosen_mode = mode
            chosen_priority = priorities[act]
    return chosen_mode

# Example usage:
print(room_light_color("reading", 20))  # Warm
print(room_light_color("sleep", 23))    # Dim
```

**Bicycle gear suggestion**

```python
def bicycle_gear_suggestion(incline_percent, speed_kmph, wind_speed_kmph):
    if incline_percent > 8:
        gear = 1 if wind_speed_kmph > 20 else 2
    elif 3 <= incline_percent <= 8:
        gear = 3 if wind_speed_kmph > 20 else 4
    elif incline_percent < 3 and speed_kmph > 25:
        gear = 7 if wind_speed_kmph < 10 else 6
    else:
        gear = 5

    # Adjust gear based on headwind and speed
    if wind_speed_kmph > 20 and speed_kmph < 20:
        gear = max(1, gear - 1)

    return gear

# Example usage:
print(bicycle_gear_suggestion(10, 15, 25)) # 1
print(bicycle_gear_suggestion(5, 20, 15))  # 3
print(bicycle_gear_suggestion(2, 30, 5))   # 7
```
